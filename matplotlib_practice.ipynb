#Import dependencies
import matplotlib.pyplot as plt

```python
#Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y_axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
#Cretae labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit between 0 and 45.
plt.ylim(0,45)
#Create a title.
plt.title("PyBer Fare by Month")
#Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
#Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create a title.
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
```

```python
#Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
#Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create a title
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
```

```python
# Create the plot-Object oriented 
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis, color="cyan", label='Chicago')
# Create labels for the x and y axes
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#Create a title
ax.set_title("PyBer Fare by Month")
#Add the legend
ax.legend()
```

```python
plt.scatter( y_axis,x_axis, color="r", label='Chicago')
plt.gca().invert_yaxis()
# Create labels for the x and y axes
plt.xlabel("Fares($)")
plt.ylabel("Date")
#Create title
plt.title("PyBer Fare By Month")
#Add the legend
plt.legend()
#Set the x limit between 0 and 45.
plt.xlim(0,45)
```

```python

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax= plt.subplots()
ax.scatter(x_axis, y_axis, color="sky blue" s=[i*5 for i in y_axis], label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")
#Create title
ax.set_title("PyBer Fare By Month")
#Add the legend
ax.legend()
#Set the x limit between 0 and 45.
ax.set_xlim(0,45)
```

```python
#Assign 12 colors, one for each month
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')
plt.show()
```

```python
# Create Pie Chart using the Object-Oriented
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
